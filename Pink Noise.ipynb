```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
#stats.set_seed(0)
```

```python
c = 0
X0 = 0
alpha = .8
steps = 1000
epsilon = stats.norm.rvs(size=steps)*1500

X = [X0]

for i in range(len(epsilon)):
    x = c + alpha*X[i] + epsilon[i]
    X.append(x)
```

```python
plt.figure(figsize=(16,4))
plt.plot(X[:], color='pink')
plt.xlabel(r"Time, t")
plt.savefig('pinknoise.png', transparent=True)
```

```python
import statsmodels.api as sm
sm.graphics.tsa.plot_acf(X, lags=40)
plt.plot()
plt.savefig('Autcorrelation_pinknoise.png', transparent=True)
```

```python
sm.graphics.tsa.plot_pacf(X, lags=40)
plt.plot()
plt.savefig('Partial_Autcorrelation_pinknoise.png', transparent=True)
```
